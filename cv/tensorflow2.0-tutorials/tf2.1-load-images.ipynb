# Load using keras.preprocessing

```python
BATCH_SIZE = 32
IMG_HEIGHT = 224
IMG_WIDTH = 224
STEPS_PER_EPOCH = np.ceil(image_count/BATCH_SIZE)
```

```python
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

```python
train_data_gen = image_generator.flow_from_directory(directory=str(data_dir),
                                                    batch_size=BATCH_SIZE,
                                                    shuffle=True,
                                                    target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                    classes=list(CLASS_NAMES))
```

## Inspect a batch

```python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10, 10))
    for n in range(25):
        ax = plt.subplot(5, 5, n+1)
        plt.imshow(image_batch[n])
        plt.title(CLASS_NAMES[label_batch[n]==1][0].title())
        plt.axis('off')
```

```python
image_batch, label_batch = next(train_data_gen)
show_batch(image_batch, label_batch)
```

```python
def get_label(file_path):
    '''从文件路径中获取label'''
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == CLASS_NAMES
```

```python
def decode_img(img):
    # 解析图片为三维uint8的张量
    img = tf.image.decode_jpeg(img, channels=3)
    # 转为[0,1]中的float类型
    img = tf.image.convert_image_dtype(img, tf.float32)
    # 重新定义图片大小
    return tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
```

```python
def process_path(file_path):
    '''获取（image, label）dataset'''
    label = get_label(file_path)
    # 从文件路径加载图片源数据
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, label
```

```python
labeled_ds = list_ds.map(process_path, num_parallel_calls=tf.data.experimental.AUTOTUNE)
```

```python
for image, label in labeled_ds.take(1):
    print('Image shape: ', image.numpy().shape)
    print('Label: ', label.numpy())
```

```python
def prepare_for_training(ds, cache=True, shuffle_buffer_size=1000):
    """此数据集小，只需一次就能全部加载到内存。当数据集过大不适合全部加载至内存时， 使用.cache(filename)缓存"""
    if cache:
        if isinstance(cache, str):
            ds = ds.cache(cache)
        else:
            ds = ds.cache()
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
    return ds
```

```python
train_ds = prepare_for_training(labeled_ds)
image_batch, label_batch = next(iter(train_ds))
```

```python
show_batch(image_batch.numpy(), label_batch.numpy())
```

```python
import time
default_timeit_steps = 1000
```

```python
def timeit(ds, steps=default_timeit_steps):
    '''检验数据集性能'''
    start = time.time()
    it = iter(ds)
    for i in range(steps):
        batch = next(it)
        if i%10 == 0:
            print('.', end='')
    print()
    end = time.time()
    duration = end - start
    print('{} batches: {} s'.format(steps, duration))
    print('{:0.5f} Images/s'.format(BATCH_SIZE*steps/duration))
```

```python
timeit(train_data_gen)
```

```python
timeit(train_ds) # 缓存到内存，速度快14倍
```

```python
uncached_ds = prepare_for_training(labeled_ds, cache=False) # 无缓存，速度快了2.5倍
timeit(uncached_ds)
```

```python
filecache_ds = prepare_for_training(labeled_ds, cache="./flowers.tfcache") # 缓存到文件，第一次速度快了3倍多
timeit(filecache_ds)
```
