```python
import tensorflow as tf
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
x_train = iris.data
y_train = iris.target
```

```python
np.random.seed(116)
np.random.shuffle(x_train)
np.random.seed(116)
np.random.shuffle(y_train)
tf.random.set_seed(116)
```

```python
# Sequential Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(3, activation='softmax', kernel_regularizer=tf.keras.regularizers.l2())
])
```

```python
# model by subclass 
class IrisModel(tf.keras.Model):
    def __init__(self):
        super(IrisModel, self).__init__()
        self.d1 = tf.keras.layers.Dense(3, activation='softmax', kernel_regularizer=tf.keras.regularizers.l2())
        
    def call(self, x):
        y = self.d1(x)
        return y
    
model_subclass  = IrisModel()
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['sparse_categorical_accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=32, epochs=500, validation_split=0.2, validation_freq=20)
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt

loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['sparse_categorical_accuracy']
val_acc = history.history['val_sparse_categorical_accuracy']

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(range(500), loss, label='Training Loss')
plt.plot(range(25), val_loss, label='Validation Loss')
plt.legend('upper median')
plt.title('Train-validation Loss')

plt.subplot(1, 2, 2)
plt.plot(range(500), acc, label='Training Acc')
plt.plot(range(25), val_acc, label='Validation Acc')
plt.legend('lower median')
plt.title('Train-validation ACC')
plt.show()
```
